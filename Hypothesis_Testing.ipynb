HYPOTHESIS TESTING

Background:  
Bombay hospitality Ltd. operates a franchise model for producing exotic Norwegian dinners throughout New England. The operating cost for a franchise in a week (W) is given by the equation W = $1,000 + $5X, where X represents the number of units produced in a week. Recent feedback from restaurant owners suggests that this cost model may no longer be accurate, as their observed weekly operating costs are higher.   
Objective:  
To investigate the restaurant owners' claim about the increase in weekly operating costs using hypothesis testing.    
Data Provided:  
●	The theoretical weekly operating cost model: W = $1,000 + $5X  
●	Sample of 25 restaurants with a mean weekly cost of Rs. 3,050   
●	Number of units produced in a week (X) follows a normal distribution with a mean (μ) of 600 units and a standard deviation (σ) of 25 units


Assignment Tasks:
1. State the Hypotheses statement:


```python
# Step 1: State the Hypotheses
# Null Hypothesis (H0):
# The mean weekly operating cost is Rs. 4000.
# H0: μ = 4000
# Alternative Hypothesis (H1):
# The mean weekly operating cost is greater than Rs. 4000.
# H1: μ > 4000
print("Null Hypothesis (H0): μ = 4000")
print("Alternative Hypothesis (H1): μ > 4000")
print("This is a right-tailed hypothesis test.")
```

2. Calculate the Test Statistic:
Use the following formula to calculate the test statistic(z):  
where:  
●	ˉxˉ = sample mean weekly cost (Rs. 3,050)  
●	μ = theoretical mean weekly cost according to the cost model (W = $1,000 + $5X for X = 600 units)  
●	σ = 5*25 units  
●	n = sample size (25 restaurants)


```python
# Given values
x_bar = 3050      # Sample mean
mu = 1000 + 5*600 # Theoretical mean
sigma = 5*25      # Population standard deviation
n = 25            # Sample size
```

```python
# Calculate the Z-test statistic
z = (x_bar - mu) / (sigma / (n ** 0.5))
print("Z-test Statistic =", z)
```

3. Determine the Probability and compare:   
Using the alpha level of 5% (α = 0.05),


```python
from scipy.stats import norm
# Significance level
alpha = 0.05
# Z-test statistic (from Task 2)
z = -38.0
# Since H1: μ > 4000 (right-tailed test)
p_value = 1 - norm.cdf(z)
print("P-value =", p_value)
```

```python
critical_z = norm.ppf(1 - alpha)
print("Critical Z-value =", critical_z)
if p_value < alpha:
    print("Reject the Null Hypothesis (H0)")
else:
    print("Fail to Reject the Null Hypothesis (H0)")
```

4. Make a Decision:  
Compare the test statistic with the critical value to decide whether to reject the null hypothesis.


```python
from scipy.stats import norm
# Significance level
alpha = 0.05
# Test statistic from Task 2
z = -38.0
# Critical value for a right-tailed test
critical_value = norm.ppf(1 - alpha)
print("Test Statistic (Z) =", z)
print("Critical Value =", critical_value)
# Decision
if z > critical_value:
    print("Decision: Reject the Null Hypothesis (H0)")
else:
    print("Decision: Fail to Reject the Null Hypothesis (H0)")
```

5. Conclusion:    
Based on the decision in step 4, conclude whether there is strong evidence to support the restaurant owners' claim that the weekly operating costs are higher than the model suggests.



```python
if z > critical_value:
    print("Conclusion:")
    print("Reject the Null Hypothesis.")
    print("There is sufficient evidence at the 5% significance level to support the restaurant owners' claim that the weekly operating costs are higher than the model suggests.")
else:
    print("Conclusion:")
    print("Fail to Reject the Null Hypothesis.")
    print("There is insufficient evidence at the 5% significance level to support the restaurant owners' claim that the weekly operating costs are higher than the model suggests.")
```

```python
# HYPOTHESIS TESTING
# Bombay Hospitality Ltd. - Weekly Operating Cost Analysis
from scipy.stats import norm
import math
# Step 1: State the Hypotheses
print("Step 1: Hypothesis Statement")
print("Null Hypothesis (H0): μ = 4000")
print("Alternative Hypothesis (H1): μ > 4000")
print("This is a Right-Tailed Z-Test.\n")
# Step 2: Given Data
x_bar = 3050            # Sample mean
mu = 1000 + (5 * 600)   # Theoretical mean
sigma = 5 * 25          # Population standard deviation
n = 25                  # Sample size
alpha = 0.05            # Significance level
print("Step 2: Given Data")
print("Sample Mean =", x_bar)
print("Theoretical Mean =", mu)
print("Population Standard Deviation =", sigma)
print("Sample Size =", n)
print()
# Step 3: Calculate the Test Statistic
z = (x_bar - mu) / (sigma / math.sqrt(n))
print("Step 3: Test Statistic")
print("Z-Test Statistic =", round(z, 4))
print()
# Step 4: Determine the Critical Value and P-value
critical_value = norm.ppf(1 - alpha)
p_value = 1 - norm.cdf(z)
print("Step 4: Critical Value and P-value")
print("Critical Value =", round(critical_value, 4))
print("P-value =", p_value)
print()
# Step 5: Decision
print("Step 5: Decision")
if z > critical_value:
    print("Decision: Reject the Null Hypothesis (H0)")
else:
    print("Decision: Fail to Reject the Null Hypothesis (H0)")
print()
# Step 6: Conclusion
print("Step 6: Conclusion")
if z > critical_value:
    print("There is sufficient evidence at the 5% significance level")
    print("to support the restaurant owners' claim that the weekly")
    print("operating costs are higher than the model suggests.")
else:
    print("There is insufficient evidence at the 5% significance level")
    print("to support the restaurant owners' claim that the weekly")
    print("operating costs are higher than the model suggests.")
```
