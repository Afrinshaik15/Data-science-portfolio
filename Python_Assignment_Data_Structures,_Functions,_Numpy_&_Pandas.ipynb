```python
'''6.	Given a 'marks' variable, print the grade:
- 90 and above: 'A'
- 75–89: 'B'
- 50–74: 'C'
- Below 50: 'Fail' '''
marks=int(input("enter your marks:"))
if marks>=90:
  print("grade:A")
elif marks<=89 and marks>=75:
  print("grade:B")
elif marks<=74 and marks>=50:
  print("grade:C")
else:
  print("Fail")
```

```python
##Section B: Numpy
'''8. Create:
- A scalar using np.array(5)
- A 1D array with values 1 to 5
- A 2D array (2x3) with values from 10 to 60 in steps of 10'''
import numpy as np
# Create a scalar
scalar_array = np.array(5)
print("Scalar array:", scalar_array)
# Create a 1D array with values 1 to 5
# np.arange(start, stop) generates numbers up to, but not including stop
array = np.arange(1, 6)
print("1D array:", array)
# Create a 2D array (2x3) with values from 10 to 60 in steps of 10
array = np.arange(10, 70, 10).reshape(2,3)
print("2D array:\n", array)
```

```python
#Section C: Exploring Pandas
'''11.	Create a small DataFrame manually with 10 rows and columns: 'Name', 'Age', 'City', and 'Salary'. Then:
- Use .info(), .describe()
- Select 'Name' and 'City' columns
- Drop 'City' column
- Fill any missing values in 'Salary' column with the mean
- Remove any duplicate rows'''
import pandas as pd
import numpy as np
data={'name':['Afrin','sajju','irfan','lahari','narendra','haritha','sadiya','divya','shabbu','shaheena','Faridha'],'Age':[22,23,47,45,35,42,44,50,20,40,55],'city':['kadapa','puttur','chennai','pulivendula','dharmavaram','kadapa','hyderbad','vizag','tirupathi','karnool','nellor'],'salary':[40000,20000,60000,None,30000,70000,40000,50000,60000,70000,80000]}
print(data)
df=pd.DataFrame(data)
print(df)
#Use .info()
a=df.info()
print(a)
#Use .describe()
b=df.describe()
print(b)
#Select 'Name' and 'City' columns
c=df[['name','city']]
print(c)
#Drop 'City' column
d=df.drop(columns=['city'])
print(d)
#Fill any missing values in 'Salary' column with the mean
e=df['salary'].mean()
print(e)
e=df['salary'].fillna(e)
print(e)
#Remove any duplicate rows
df.duplicated()
print(df.duplicated())
```
