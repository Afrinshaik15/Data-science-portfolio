●	Discuss the scenarios where each scaling technique is preferred and why. **Answer:**                         1. Standard Scaling (Z-score Scaling)  
Standard Scaling transforms the data so that it has a mean of 0 and a standard deviation of 1.   
Preferred Scenarios:     
When the data follows a normal (Gaussian) distribution.  
For algorithms that are sensitive to the scale of features, such as:  
Logistic Regression  
Support Vector Machine (SVM)
K-Nearest Neighbors (KNN)
Principal Component Analysis (PCA)
Neural Networks  
**Why?**  
It standardizes features to a common scale without restricting the range.
It improves model convergence and performance for distance- and gradient-based algorithms.    
2. Min-Max Scaling (Normalization)   
Min-Max Scaling rescales the data to a fixed range, usually 0 to 1.   
Preferred Scenarios:   
When features have different ranges and need to be normalized.  
For algorithms that require inputs within a bounded range, such as:   
Neural Networks
K-Nearest Neighbors (KNN)
Deep Learning models  
Why?   
It preserves the original distribution of the data while scaling it.
It is useful when the minimum and maximum values are known and there are no significant outliers.

●	Discuss the pros and cons of One-Hot Encoding and Label Encoding.

**One-Hot Encoding**     
**Pros:**   
Does not introduce any ordinal relationship between categories.   
Suitable for nominal (unordered) categorical variables.   
Improves performance for many machine learning algorithms that cannot interpret categorical values directly.   
**Cons:**   
Increases the number of features, leading to higher memory usage.  
Can create a sparse dataset when there are many unique categories.
May increase training time for large datasets.  
**Label Encoding**   
**Pros:**  
Simple and easy to implement.   
Does not increase the number of features.   
Efficient in terms of memory and computation.  
Suitable for ordinal (ordered) categorical variables and tree-based models.  
**Cons:**   
Assigns numerical values to categories, which may create a false ordinal relationship.   
Can negatively affect algorithms like Linear Regression, Logistic Regression, KNN, and SVM if applied to nominal data.    
Less suitable for nominal categorical variables with no natural ordering.

```python
# Create Age Group feature
df['age_group'] = pd.cut(df['age'],bins=[0, 25, 45, 65, 100],labels=['Young', 'Adult', 'Middle_Aged', 'Senior'])
df[['age', 'age_group']].head()
```

**Rationale:**   
Grouping individuals by age captures different career stages and earning potential, which may improve income prediction.

```python
# Create Net Capital feature
df['net_capital'] = df['capital_gain'] - df['capital_loss']
df[['capital_gain', 'capital_loss', 'net_capital']].head()
```

**Rationale:**   
Combining capital-gain and capital-loss into a single feature provides the overall net capital impact, which can be a stronger predictor of whether an individual's income exceeds $50K.

●	Apply a transformation (e.g., log transformation) to at least one skewed numerical feature and justify your choice.

```python
# Check skewness
df.select_dtypes(include=['int64', 'float64']).skew()
```

```python
# Apply log transformation
df['capital_gain_log'] = np.log1p(df['capital_gain'])
# Display transformed feature
df[['capital_gain', 'capital_gain_log']].head()
```

**Justification**:  
The capital-gain feature is highly right-skewed, with many zero values and a few very large values.   
Applying log1p() reduces skewness, minimizes the effect of extreme values, and makes the distribution more suitable for machine learning algorithms.   
log1p() is used instead of log() because it safely handles zero values (log1p(0) = 0).
